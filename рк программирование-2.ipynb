```python
import pandas as pd
```

```python
df = pd.read_csv("global_earthquakes_10yrs_M4.5_2025-12-10.csv")
df.head()
```

```python
print(df.info())
df.describe(include="all").T
```

```python

time_col = "time"

df[time_col] = pd.to_datetime(df[time_col], errors="coerce")


df[time_col].head()
```

```python
df["year"] = df[time_col].dt.year
df["month"] = df[time_col].dt.month
df["day"] = df[time_col].dt.day
df["hour"] = df[time_col].dt.hour
df["date"] = df[time_col].dt.date

# Магнитудные бины
df["mag_bin"] = pd.cut(
    df["mag"], 
    bins=[4.5,5,5.5,6,6.5,7,10],
    labels=["4.5–5","5–5.5","5.5–6","6–6.5","6.5–7","7+"],
    right=False
)

# Глубина
df["depth_km"] = pd.to_numeric(df["depth"], errors="coerce")

# Извлекаем "страну" из place
def extract_country(place):
    if pd.isna(place):
        return np.nan
    parts = [p.strip() for p in str(place).split(",")]
    if len(parts) > 1:
        return parts[-1]
    return place

df["country"] = df["place"].apply(extract_country)
```

```python
# Дубликаты
before = len(df)
df = df.drop_duplicates()
after = len(df)

print("Удалено дубликатов:", before - after)

# Пропуски в ключевых колонках
df = df.dropna(subset=["latitude", "longitude", "mag", time_col])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
monthly = df.groupby(pd.Grouper(key="time", freq="M")).size()

plt.figure(figsize=(12,4))
plt.plot(monthly.index, monthly.values)
plt.title("Earthquakes per Month")
plt.xlabel("Month")
plt.ylabel("Count")
plt.grid(True)
plt.savefig("plots/01_monthly_counts.png")
plt.show()
```

```python
yearly = df.groupby("year").size()

plt.figure(figsize=(8,4))
yearly.plot(kind="bar")
plt.title("Earthquakes per Year")
plt.savefig("plots/02_yearly_counts.png")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(df["depth_km"], bins=40)
plt.title("Depth Distribution")
plt.xlabel("Depth (km)")
plt.savefig("plots/04_depth_hist.png")
plt.show()
```
